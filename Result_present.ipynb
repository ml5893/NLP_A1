```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pkl.load(open("result/val_acc_vs_step_withPunc.p","rb"))
data0 = pkl.load(open("result/val_acc_vs_step_BaseModel.p", "rb"))
_ , r1 = zip(*data1)
step, r0 = zip(*data0)

plt.plot(step, r0, label='no punc')
plt.plot(step, r1, label='punc')
plt.legend()
plt.title('tokenization with or without punctuations')
plt.savefig("textfile/com1.pdf")
plt.show()
```

```python
data0 = pkl.load(open("result/val_acc_vs_step_BaseModel.p", "rb"))
data1 = pkl.load(open("result/val_acc_vs_step_n1.p","rb"))
data2 = pkl.load(open("result/val_acc_vs_step_n2.p","rb"))
data3 = pkl.load(open("result/val_acc_vs_step_n3.p","rb"))
data4 = pkl.load(open("result/val_acc_vs_step_n4.p","rb"))
data3a = pkl.load(open("result/val_acc_vs_step_n3a.p","rb"))
data4a = pkl.load(open("result/val_acc_vs_step_n4a.p","rb"))

step, r0 = zip(*data0)
_ , r1 = zip(*data1)
_ , r2 = zip(*data2)
_ , r3 = zip(*data3)
_ , r4 = zip(*data4)
_ , r3a = zip(*data3a)
_ , r4a = zip(*data4a)

plt.plot(step, r0, label='base')
plt.plot(step, r1, label='n=1')
plt.plot(step, r2, label='n=2')
plt.plot(step, r3, label='n=3')
plt.plot(step, r4, label='n=4')
plt.plot(step, r3a, label='n=3(a)')
plt.plot(step, r4a, label='n=4(a)')
plt.legend()
plt.xlabel('step')
plt.ylabel('val acc')
plt.title('n-gram')
plt.savefig("textfile/com2.pdf")
plt.show()
```

```python
data0 = pkl.load(open("result/val_acc_vs_step_BaseModel.p", "rb"))
data1 = pkl.load(open("result/val_acc_vs_step_vs1.p","rb"))
data2 = pkl.load(open("result/val_acc_vs_step_vs2.p","rb"))
data3 = pkl.load(open("result/val_acc_vs_step_vs3.p","rb"))

step, r0 = zip(*data0)
_ , r1 = zip(*data1)
_ , r2 = zip(*data2)
_ , r3 = zip(*data3)

plt.plot(step, r0, label='base')
plt.plot(step, r1, label='5000')
plt.plot(step, r2, label='10000')
plt.plot(step, r3, label='20000')
plt.legend()
plt.xlabel('step')
plt.ylabel('val acc')
plt.title('vocabulary size')
plt.savefig("textfile/com3.pdf")
plt.show()
```

```python
data0 = pkl.load(open("result/val_acc_vs_step_BaseModel.p", "rb"))
data1 = pkl.load(open("result/val_acc_vs_step_SGD.p", "rb"))

step, r0 = zip(*data0)
_, r1 = zip(*data1)

plt.plot(step, r0, label='base')
plt.plot(step, r1[:len(step)], label='SGD')
plt.legend()
plt.xlabel('step')
plt.ylabel('val acc')
plt.title('optimizer')
plt.savefig("textfile/com5.pdf")
plt.show()
```

```python
data0 = pkl.load(open("result/val_acc_vs_step_BaseModel.p", "rb"))
data1 = pkl.load(open("result/val_acc_vs_step_lr1.p","rb"))
data2 = pkl.load(open("result/val_acc_vs_step_lr2.p","rb"))
data3 = pkl.load(open("result/val_acc_vs_step_lrdecay.p","rb"))
data4 = pkl.load(open("result/val_acc_vs_step_lrdecay2.p","rb"))

step, r0 = zip(*data0)
_ , r1 = zip(*data1)
_ , r2 = zip(*data2)
_ , r3 = zip(*data3)
_ , r4 = zip(*data4)
```

```python
plt.plot(step, r0, label='base')
plt.plot(step, r1, label='lr=0.01')
plt.plot(step, r2, label='lr=0.0001')
plt.plot(step, r3, label='lr=0.001*(0.8)^epoch')
plt.plot(step, r4, label='lr=0.002/epoch')

plt.legend()
plt.xlabel('step')
plt.ylabel('val acc')
plt.title('learning rate')
plt.savefig("textfile/com6.pdf")
plt.show()
```
